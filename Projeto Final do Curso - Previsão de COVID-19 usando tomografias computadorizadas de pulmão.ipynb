## 1. Entendimento dos Dados

 O monitoramento da COVID-19 usando tomografias computadorizadas (TC) de pulmão tem sido uma área de pesquisa de suma importância desde o início da pandemia. A TC de pulmão pode fornecer informações valiosas sobre o envolvimento pulmonar e a gravidade da infecção. Aqui estão alguns pontos-chave sobre o uso de TC para prever e avaliar a COVID-19:
 
Detecção e Diagnóstico: A TC pode detectar padrões característicos de infecção por COVID-19, como opacidades em vidro fosco, consolidações e padrões em “vidro fosco” que são comuns em casos mais graves. Monitoramento da Progressão: A TC de pulmão pode ser usada para monitorar a evolução da doença ao longo do tempo, ajudando a ajustar o tratamento conforme necessário. Mudanças nos padrões pulmonares ao longo do tempo podem fornecer insights sobre a resposta ao tratamento.
Modelos de Previsão: Pesquisadores têm desenvolvido modelos de aprendizado de máquina e inteligência artificial que analisam imagens de TC para prever o risco de complicações, como necessidade de ventilação mecânica ou hospitalização prolongada. Esses modelos treinam algoritmos para identificar características associadas a desfechos adversos.

Limitações e Considerações: Embora a TC seja uma ferramenta poderosa, ela não substitui os testes laboratoriais para diagnóstico definitivo e avaliação completa. Além disso, a exposição à radiação é uma preocupação, especialmente com a realização frequente de exames.
Integração com Outros Dados: A combinação de informações de TC com dados clínicos, laboratoriais e históricos de saúde pode melhorar a acurácia das previsões e a tomada de decisões clínicas.

O uso de TC de pulmão para COVID-19 é um campo em evolução, e a pesquisa continua a refinar as técnicas e modelos para melhorar a precisão e a utilidade clínica.

## 2. Arquitetura Utilizada

A arquitetura escolhida para a realização da previsão de COVID-19 TC DE Pulmões foi a VGG16. 

A arquitetura VGG16 é um modelo de rede neural convolucional (CNN) que se destacou em tarefas de reconhecimento de imagem e outros problemas de visão computacional. Desenvolvida pelo Visual Geometry Group da Universidade de Oxford, o VGG16 é conhecido por sua simplicidade e eficácia. Aqui está uma explicação detalhada da arquitetura:

<u>Estrutura Geral:</u>

O VGG16 é uma rede neural profunda que possui 16 camadas treináveis, que incluem camadas convolucionais e camadas totalmente conectadas (ou densas). A arquitetura é projetada para ser simples, com um padrão de camadas repetitivas, o que facilita a compreensão e a implementação.
Camadas Convolucionais

A rede é composta principalmente por camadas convolucionais seguidas de camadas de pooling (redução de dimensionalidade). Essas camadas são responsáveis por extrair características das imagens:

Camadas Convolucionais: São responsáveis pela extração de características locais da imagem. O VGG16 utiliza filtros 3x3, que são pequenos e permitem a construção de características complexas de forma hierárquica. Cada camada convolucional é seguida por uma função de ativação ReLU (Rectified Linear Unit), que introduz não-linearidade na rede.

Camadas de Pooling: Após as camadas convolucionais, são aplicadas camadas de max pooling com uma janela de 2x2 e um stride de 2. O max pooling reduz a dimensionalidade das características extraídas e ajuda a manter as características mais importantes enquanto diminui o tamanho da representação.

<u>Camadas da VGG16<u/>

A arquitetura do VGG16 segue um padrão de repetição das camadas convolucionais e de pooling. Aqui está uma visão geral da estrutura detalhada:
Camadas Convolucionais:
Camadas 1: 2 camadas convolucionais com 64 filtros 3x3.
Camadas 2: 2 camadas convolucionais com 128 filtros 3x3.
Camadas 3: 3 camadas convolucionais com 256 filtros 3x3.
Camadas 4: 3 camadas convolucionais com 512 filtros 3x3.
Camadas 5: 3 camadas convolucionais com 512 filtros 3x3.

Camadas de Pooling: Após cada conjunto de camadas convolucionais, uma camada de max pooling é aplicada.

Camadas Densas:

Camadas 6 e 7: 2 camadas densas (fully connected) com 4096 neurônios cada.
Camada 8: Uma camada densa com 1000 neurônios para a classificação final (no caso da arquitetura original, para 1000 classes do ImageNet).

<u>Saída da Rede</u>

Após as camadas densas, a rede termina com uma camada de softmax para obter as probabilidades de cada classe na tarefa de classificação, com base na saída das camadas densas.

<u>Características Importantes</u>

Profundidade: A arquitetura é relativamente profunda, com 16 camadas treináveis, o que permite capturar características complexas e de alto nível das imagens.

Simetria e Repetição: O uso repetitivo de filtros 3x3 e max pooling facilita o design e a implementação, e também ajuda na construção de representações hierárquicas de características.

Transferência de Aprendizado: O VGG16 é frequentemente usado como base para transferência de aprendizado, onde suas camadas convolucionais pré-treinadas em grandes conjuntos de dados, como o ImageNet, são reutilizadas para novos problemas de visão computacional.

<u>Aplicações</u>

A VGG16 tem sido amplamente utilizada em várias tarefas de visão computacional, incluindo:

- Classificação de Imagens: Identificar e classificar objetos em imagens.
- Detecção de Objetos: Identificar e localizar objetos em uma imagem.
- Segmentação de Imagens: Dividir imagens em segmentos com base em características específicas.

Em resumo, o VGG16 é um modelo de CNN que se destaca por sua simplicidade e profundidade, oferecendo um bom equilíbrio entre complexidade e desempenho. Sua arquitetura é fácil de entender e implementar, tornando-o uma escolha popular para muitos problemas de visão computacional.

## Processamento dos dados com Data Augmentation utilizando ImageDataGenerator 

```python
##Definindo o diretório raiz
```

```python
diretorio_raiz = r"C:\Users\beatriz.itacaramby\Documents\Projeto Final"
```

```python
# Define o número desejado de imagens em cada pasta
num_imagens_desejadas = 500

# Calcula o peso de cada classe para equilibrar o número de imagens
class_weights = {}
total_images = 0

for classe in os.listdir(diretorio_raiz):
    num_imagens = len(os.listdir(os.path.join(diretorio_raiz, classe)))
    class_weights[classe] = num_imagens_desejadas / num_imagens if num_imagens != 0 else 1.0
    total_images += num_imagens

# Normaliza os pesos das classes
total_classes = len(class_weights)
total_weights = sum(class_weights.values())
for classe, weight in class_weights.items():
    class_weights[classe] = (weight * total_classes) / total_weights

    
datagen = ImageDataGenerator(
    horizontal_flip=True,  # Aplica espelhamento horizontal
    rotation_range=20,      # Define o intervalo de rotação de 20 graus
    rescale=1./255          # Normaliza os valores dos pixels
)

# Carrega e pré-processa as imagens do diretório especificado usando o gerador de dados em lote
train_generator = datagen.flow_from_directory(
    directory=diretorio_raiz,
    target_size=(224, 224),  # Redimensiona todas as imagens para 224x224 pixels
    batch_size=32,  # Define o tamanho do lote como o número desejado de imagens
    class_mode='categorical'  # Define o modo de classe para problemas de classificação multiclasse
)

# Calcula o número de épocas com base no número total de imagens e o tamanho do lote
num_epocas = total_images // num_imagens_desejadas


total_images
num_imagens_desejadas
num_epocas
```

```python
class_indices = train_generator.class_indices

# Obtém as classificações das imagens
class_labels = list(class_indices.keys())
class_counts = train_generator.classes

# Calcula a proporção de cada classe
class_proportions = {label: count / len(class_counts) for label, count in zip(class_labels, np.bincount(class_counts))}

# Exibe as classificações e suas proporções
print("Classificações das Imagens:")
for label, proportion in class_proportions.items():
    print(f"{label}: {proportion:.2%}")
```

## Modelo VGG16

```python
from keras import models
from keras.layers import Conv2D, MaxPool2D, Flatten, Dense, Input

# Definindo o modelo Sequential
model = models.Sequential()

# Adicionando a camada de entrada
model.add(Input(shape=(224, 224, 3)))  # Usando Input(shape) como a primeira camada

# Primeira camada convolucional
model.add(Conv2D(filters=64, kernel_size=(3, 3), padding="same", activation="relu"))
model.add(Conv2D(filters=64, kernel_size=(3, 3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2, 2), strides=(2, 2)))

# Segunda camada convolucional
model.add(Conv2D(filters=128, kernel_size=(3, 3), padding="same", activation="relu"))
model.add(Conv2D(filters=128, kernel_size=(3, 3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2, 2), strides=(2, 2)))

# Terceira camada convolucional
model.add(Conv2D(filters=256, kernel_size=(3, 3), padding="same", activation="relu"))
model.add(Conv2D(filters=256, kernel_size=(3, 3), padding="same", activation="relu"))
model.add(Conv2D(filters=256, kernel_size=(3, 3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2, 2), strides=(2, 2)))

# Quarta camada convolucional
model.add(Conv2D(filters=512, kernel_size=(3, 3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3, 3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3, 3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2, 2), strides=(2, 2)))

# Quinta camada convolucional
model.add(Conv2D(filters=512, kernel_size=(3, 3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3, 3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3, 3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2, 2), strides=(2, 2)))

# Adicionando camadas densas
model.add(Flatten())
model.add(Dense(units=4096, activation="relu"))
model.add(Dense(units=4096, activation="relu"))
model.add(Dense(units=2, activation="softmax"))

# Compilando o modelo
model.compile(optimizer='adam', 
              loss='categorical_crossentropy',  
              metrics=['accuracy'])
```

```python
# Compilação do Modelo utilizando o otimizador Adam
opt = Adam(learning_rate=0.001, beta_1=0.5)
model.compile(optimizer=opt, loss=keras.losses.categorical_crossentropy, metrics=['accuracy'])
```
